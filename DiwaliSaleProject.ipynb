```python
sns.countplot(x='Sex',data=df)
```

```python
ax=sns.countplot(x='Sex',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Sex'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
gen_sales=df.groupby(['Sex'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Sex",y="Amount",data=gen_sales)
```

```python
#from above graphs we can see Gender and purchase amount of female is greater than male
```

```python
sns.countplot(x='Age Group',data=df,hue='Sex')
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Sex')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Age_sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Age_sales)
```

```python
#from above chat we found that from age group 26-35  Female are higest buyer.
```

```python
state_sales=df.groupby(['State','Sex'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=state_sales,x='State',y='Amount',hue='Sex')
```

```python
state_sales=df.groupby(['State','Sex'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=state_sales,x='State',y='Orders',hue='Sex')
```

```python
#from above uttarprdesh there are more order placed and also more amout spend.
```
